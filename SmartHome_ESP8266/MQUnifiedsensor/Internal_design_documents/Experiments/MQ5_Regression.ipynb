```python
x_MQ5 = sheetMQ5.col_values(0)[2:]
MQ5_H2 = sheetMQ5.col_values(1)[2:]
MQ5_LPG = sheetMQ5.col_values(2)[2:]
MQ5_CH4 = sheetMQ5.col_values(3)[2:]
MQ5_CO = sheetMQ5.col_values(4)[2:]
```

```python
def zero_to_nan(values):
    """Replace every 0 with 'nan' and return a copy."""
    return [float('nan') if x==0 else x for x in values]
```

```python
MQ5_H2 =zero_to_nan(MQ5_H2)
MQ5_LPG =zero_to_nan(MQ5_LPG)
MQ5_CH4 =zero_to_nan(MQ5_CH4)
MQ5_CO =zero_to_nan(MQ5_CO)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
#from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from sklearn import linear_model

dataH2 = {'RsRo': x_MQ5, 'H2': MQ5_H2}
dataLPG = {'RsRo': x_MQ5, 'LPG': MQ5_LPG}
dataCH4 = {'RsRo': x_MQ5, 'CH4': MQ5_CH4}
dataCO = {'RsRo': x_MQ5, 'CO': MQ5_CO}

dfMQ5_H2 = pd.DataFrame(dataH2)
dfMQ5_LPG = pd.DataFrame(dataLPG)
dfMQ5_CH4 = pd.DataFrame(dataCH4)
dfMQ5_CO = pd.DataFrame(dataCO)

dfMQ5_H2['H2'] = pd.to_numeric(dfMQ5_H2['H2'])
dfMQ5_LPG['LPG'] = pd.to_numeric(dfMQ5_LPG['LPG'])
dfMQ5_CH4['CH4'] = pd.to_numeric(dfMQ5_CH4['CH4'])
dfMQ5_CO['CO'] = pd.to_numeric(dfMQ5_CO['CO'])

dfMQ5_H2['H2'] = dfMQ5_H2['H2'].replace('',None, regex=True)
dfMQ5_LPG['LPG'] = dfMQ5_LPG['LPG'].replace('',None, regex=True)
dfMQ5_CH4['CH4'] = dfMQ5_CH4['CH4'].replace('',None, regex=True)
dfMQ5_CO['CO'] = dfMQ5_CO['CO'].replace('',None, regex=True)

#Global X_Predict variable
X_Predict = dfMQ5_LPG.RsRo.apply(lambda x: [x]).tolist()
```

```python
#Model and train H2
dataset2TrainH2 = dfMQ5_H2.copy()
dataset2TrainH2.dropna(inplace=True)
X_trainH2 = dataset2TrainH2.RsRo.apply(lambda x: [x]).tolist()
y_trainH2 = dataset2TrainH2['H2'].tolist()
model = linear_model.Lasso(alpha=0.1)
model.fit(X_trainH2, y_trainH2)
#Predict
H2_Predicted = model.predict(X_Predict)
#save into MQ2
MQ5_H2 = H2_Predicted
```

```python
#Model and train LPG
dataset2TrainLPG = dfMQ5_LPG.copy()
dataset2TrainLPG.dropna(inplace=True)
X_trainLPG = dataset2TrainLPG.RsRo.apply(lambda x: [x]).tolist()
y_trainLPG = dataset2TrainLPG['LPG'].tolist()
model = linear_model.Lasso(alpha=0.1)
model.fit(X_trainLPG, y_trainLPG)
#Predict
LPG_Predicted = model.predict(X_Predict)
#save into MQ2
MQ5_LPG = LPG_Predicted
```

```python
#Model and train CH4
dataset2TrainCH4 = dfMQ5_CH4.copy()
dataset2TrainCH4.dropna(inplace=True)
X_trainCH4 = dataset2TrainCH4.RsRo.apply(lambda x: [x]).tolist()
y_trainCH4 = dataset2TrainCH4['CH4'].tolist()
model = linear_model.Lasso(alpha=0.1)
model.fit(X_trainCH4, y_trainCH4)
#Predict
CH4_Predicted = model.predict(X_Predict)
#save into MQ2
MQ5_CH4 = CH4_Predicted
```

```python
#Model and train CO
dataset2TrainCO = dfMQ5_CO.copy()
dataset2TrainCO.dropna(inplace=True)
X_trainCO = dataset2TrainCO.RsRo.apply(lambda x: [x]).tolist()
y_trainCO = dataset2TrainCO['CO'].tolist()
model = linear_model.Lasso(alpha=0.1)
model.fit(X_trainCO, y_trainCO)
#Predict
CO_Predicted = model.predict(X_Predict)
#save into MQ2
MQ5_CO = CO_Predicted
```
