```python
import xlrd
book = xlrd.open_workbook("Experimento 3 - 1 de Junio.xlsx")
sheet = book.sheet_by_name("Experimento 3 - 1 de Junio")
```

```python
for row_index in range(5,20): #reading first columns
    id, MQ2_R0, MQ3_R0, MQ4_R0, MQ5_R0, MQ6_R0, MQ7_R0, MQ8_R0, MQ9_R0, MQ2_Vm, MQ3_Vm, MQ4_Vm, MQ5_Vm, MQ6_Vm, MQ7_Vm, MQ8_Vm, MQ9_Vm = sheet.row_values(row_index)
    print(id, MQ2_R0, MQ3_R0, MQ4_R0, MQ5_R0, MQ6_R0, MQ7_R0, MQ8_R0, MQ9_R0)
```

```python
x = sheet.col_values(0)[5:]
MQ2_R0 = sheet.col_values(1)[5:]
MQ3_R0 = sheet.col_values(2)[5:]
MQ4_R0 = sheet.col_values(3)[5:]
MQ5_R0 = sheet.col_values(4)[5:]
MQ6_R0 = sheet.col_values(5)[5:]
MQ7_R0 = sheet.col_values(6)[5:]
MQ8_R0 = sheet.col_values(7)[5:]
MQ9_R0 = sheet.col_values(8)[5:]
```

```python
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import matplotlib.transforms as mtransforms

fig, ax = plt.subplots()

fig.set_size_inches(18.5, 10.5, forward=True)
fig.set_dpi(200)

# only these two lines are calibration curves
plt.plot(x,MQ2_R0, marker='o', linewidth=1, label='MQ2')
plt.plot(x, MQ3_R0, marker='o', linewidth=1, label='MQ3')
plt.plot(x, MQ4_R0, marker='o', linewidth=1, label='MQ4')
plt.plot(x, MQ5_R0, marker='o', linewidth=1, label='MQ5')
plt.plot(x, MQ6_R0, marker='o', linewidth=1, label='MQ6')
plt.plot(x, MQ7_R0, marker='o', linewidth=1, label='MQ7')
plt.plot(x, MQ8_R0, marker='o', linewidth=1, label='MQ8')
plt.plot(x, MQ9_R0, marker='o', linewidth=1, label='MQ9')

# reference line, legends, and axis labels
#line = mlines.Line2D([0, 1], [0, 1], color='black')
#transform = ax.transAxes
#line.set_transform(transform)
#ax.add_line(line)
fig.suptitle('Calibration plot for MQ data')
ax.set_xlabel('Samples')
ax.set_ylabel('R0 (Kohms)')
plt.legend()
plt.savefig('demo1.svg')
plt.show()
```
